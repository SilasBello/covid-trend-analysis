```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Make plt display inside notebook
%matplotlib inline

# For a clean and readable
sns.set_theme(style="whitegrid")
```

```python
# Load csv file
df = pd.read_csv("owid-covid-data.csv")

# show first few rows
df.head()

```

```python
# Keep only Nigeria and United State Data
filtered_df = df[df["location"].isin(["Nigeria", "United States"])].copy()

# Select only important columns
columns_to_keep = ["location", "date", "total_cases", "new_cases", "total_deaths", "new_deaths"]
filtered_df = filtered_df[columns_to_keep].copy()

# Show first few rows
filtered_df.head(100)


```

```python
# Convert "date" column to datetime format
filtered_df["date"] = pd.to_datetime(filtered_df["date"]).copy()

# Check for missing values
filtered_df.isnull().sum()
```

```python
# Fill Missing Values in "new_cases" with 0
filtered_df["new_cases"] = filtered_df["new_cases"].fillna(0)
```

```python
import seaborn as sns

sns.set_theme(style="whitegrid")

# Filter data for Nigeria and United States
nigeria_df = filtered_df[filtered_df["location"] == "Nigeria"].copy()
us_df = filtered_df[filtered_df["location"] == "United States"].copy()

# Plot total cases over time(NG vs US)
plt.figure(figsize=(14, 6))
sns.lineplot(data= nigeria_df, x="date", y="total_cases", label="Nigeria", color='green')
sns.lineplot(data= us_df, x="date", y="total_cases", label="United States", color='red')
plt.title("Total COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
# New Daily Cases Over Time

plt.figure(figsize=(14, 6))
sns.lineplot(data= nigeria_df, x= "date", y= "new_cases", label= 'Nigeria', color= 'blue')
sns.lineplot(data= us_df, x= "date", y= "new_cases", label= 'Nigeria', color= 'orange')
plt.title("New Daily COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data= nigeria_df, x= "date", y= "new_cases", label= 'Nigeria', color= 'blue')
plt.title("New Daily COVID-19 Cases Over Time in Nigeria")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Add 7 days smoothed newcases

nigeria_df["smoothed_new_cases"] = nigeria_df["new_cases"].rolling(window=7).mean()
us_df["smoothed_new_cases"] = us_df["new_cases"].rolling(window=7).mean()
```

```python
# Plot the Smoothed New cases

plt.figure(figsize=(14,6))
sns.lineplot(data= nigeria_df, x= "date", y= "smoothed_new_cases", label='Nigeria', color= 'blue')
sns.lineplot(data= us_df, x= "date", y= "smoothed_new_cases", label= 'United States', color= 'orange')
plt.title("Smoothed New Daily COVID-19 Daily Cases(7 Days Rolling Average)")
plt.xlabel("Date")
plt.ylabel("New Cases(7-day Rolling Average)")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Smooth new_deaths column
nigeria_df["smoothed_new_deaths"] = nigeria_df["new_deaths"].rolling(window=7).mean()
us_df["smoothed_new_deaths"] = us_df["new_deaths"].rolling(window=7).mean()
```

```python
# Plot
plt.figure(figsize=(14,6))
sns.lineplot(data= nigeria_df, x= "date", y= "smoothed_new_deaths", label= 'Nigeria', color= 'blue')
sns.lineplot(data= us_df, x= "date", y= "smoothed_new_deaths", label= 'United States', color= 'orange')
plt.title("Smoothed Daily COVID-19 Deaths(Nigeria vs United States)")
plt.xlabel("Date")
plt.ylabel("Smoothed Daily Deaths")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot for Total Deaths Over Time(Nigeria vs US)

plt.figure(figsize=(14,6))
sns.lineplot(data= nigeria_df, x= "date", y= "total_deaths", label= 'Nigeria', color='green')
sns.lineplot(data= us_df, x= "date", y= "total_deaths", label= "United States", color= 'red')
plt.title("Total COVID-19 Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Total Deaths")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculating cfr
nigeria_df["cfr"] = np.where(nigeria_df["total_cases"] > 0, 
                             (nigeria_df["total_deaths"]/nigeria_df["total_cases"] * 100), 
                             np.nan
                            )
us_df["cfr"] = np.where(us_df["total_cases"] > 0,
                       (us_df["total_deaths"]/us_df["total_cases"] * 100),
                        np.nan
                       )
```

```python
# Plot cfr over time

plt.figure(figsize=(14, 6))
sns.lineplot(data= nigeria_df, x= "date", y= "cfr", label= 'Nigeria', color= 'green')
sns.lineplot(data= us_df, x= "date", y= "cfr", label= 'United States', color= 'red')
plt.title("COVID-19 Case Fatality Rate Over Time")
plt.xlabel("Date")
plt.ylabel("CFR(%)")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot for case per million
# Get population value for each country 
nigeria_population = df[(df["location"] == "Nigeria") & (df["population"].notna())]["population"].iloc[-1]
us_population = df[(df["location"] == "United States") & (df["population"].notna())]["population"].iloc[-1]

# Add population column to each df
nigeria_df["population"] = nigeria_population
us_df["population"] = us_population

# Calculation Cases per million
nigeria_df["cases_per_million"] = (nigeria_df["total_cases"] / nigeria_df["population"]) * 1000000
us_df["cases_per_million"] = (us_df["total_cases"] / us_df["population"]) * 1000000

# Plot
plt.figure(figsize=(14,6))
sns.lineplot(data= nigeria_df, x= "date", y= "cases_per_million", label= 'Nigeria', color= 'green')
sns.lineplot(data= us_df, x= "date", y= "cases_per_million", label= 'United States', color= 'red')
plt.title("Total COVID-19 Cases Per Million Person Over Time")
plt.xlabel("Date")
plt.ylabel("Cases per Million")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Get latest entry for each country
latest_nigeria = nigeria_df[nigeria_df["date"] == nigeria_df["date"].max()]
latest_us = us_df[us_df["date"] == us_df["date"].max()]

# Build summary dataframe
summary_df = pd.DataFrame({
    "Country": ["Nigeria", "United States"],
    "Population (Millions)": [nigeria_population / 1_000_000, us_population / 1_000_000],
    "Cases per Million": [
        latest_nigeria["cases_per_million"].values[0],
        latest_us["cases_per_million"].values[0]
    ]
})

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=summary_df, x="Country", y="Cases per Million", palette=["green", "red"], hue='Country', legend= False)

plt.title("COVID-19 Cases per Million vs Country Population")
plt.ylabel("Total Cases per Million")
plt.xlabel("Country")
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()

```
